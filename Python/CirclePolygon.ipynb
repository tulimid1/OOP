```python
import matplotlib.pyplot as plt 
plt.rcParams.update({'font.size': 16}) 
import numpy as np 
```

# Circle 

```python
class Circle:

    def __init__(self, positionX, positionY, rad):
        self.posX = positionX
        self.posY = positionY
        self.radius = rad
        self.t = np.linspace(0, 1, 100)

        self.circle_plot()

    def __str__(self):
        return f'Your circle centered at {self.posX:.2f},{self.posY:.2f} has radius of {self.radius:.2f} with an area of {self.calculateArea():.2f} and a circumfrence of {self.calculateCircumfrence():.2f}.'

    def circle_plot(self):
        plt.figure()
        self.updateVisualization()

    def updateVisualization(self):
        xdata = np.cos(2*np.pi*self.t)*self.radius + self.posX
        ydata = np.sin(2*np.pi*self.t)*self.radius + self.posY
        plt.plot(xdata, ydata, lw=2)
        plt.xlabel('X')
        plt.ylabel('Y')
        plt.axis('equal')
        plt.show()

    def calculateArea(self):
        area = np.pi * self.radius**2
        return area

    def calculateCircumfrence(self):
        circumfrence = 2 * np.pi * self.radius
        return circumfrence
    
```

## Circle 1

```python
my_circle = Circle(0, 0, 5)
```

```python
print(my_circle)
```

```python
my_circle.posX = 5
my_circle.circle_plot()
```

```python
print(my_circle)
```

# Polygon

```python
class polygon:
    def __init__(self,positionX, positionY, length, number_verticies, rotation_angle):
        self.posX = positionX
        self.posY = positionY
        self.rad = length
        self.numV = number_verticies
        self.rotAng = rotation_angle

        self.t = np.linspace(0,1,self.numV+1)

        self.polyplot()

    def __str__(self):
        return f'Your polygon has {self.numV:.0f} sides with a perimeter of {self.calculatePerimeter():.2f}'

    def polyplot(self):
        plt.figure()
        self.updateVisualization()

    def updateVisualization(self):
        xdata = np.cos(2*np.pi*self.t+self.rotAng)*self.rad + self.posX
        ydata = np.sin(2*np.pi*self.t+self.rotAng)*self.rad + self.posY
        plt.plot(xdata, ydata, lw=2)
        plt.xlabel('X')
        plt.ylabel('Y')
        plt.axis('equal')
        plt.show()

    def calculatePerimeter(self):
        perimeter = self.numV * self.rad
        return perimeter
```

```python
my_square = polygon(0,0,4,4,90)
```

```python
print(my_square)
```

```python
my_octogon = polygon(0,0, 2, 8, 0)
```

```python
print(my_octogon)
```
